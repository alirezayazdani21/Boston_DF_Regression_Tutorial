```python
from sklearn import linear_model
```

```python
bos_X_train = boston_train.drop(columns='medv')
```

```python
bos_y_train = boston_train['medv']
```

```python
bos_X_test = boston_test.drop(columns='medv')
```

```python
bos_y_test = boston_test['medv']
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
my_skl_bos_ols=regr.fit(bos_X_train, bos_y_train)
```

```python
bos_ols_train_pred = regr.predict(bos_X_train)
```

```python
bos_ols_test_pred = regr.predict(bos_X_test)
```

```python
[metrics.r2_score(bos_y_train, bos_ols_train_pred),
 metrics.mean_squared_error(bos_y_train, bos_ols_train_pred),
 metrics.mean_absolute_error(bos_y_train, bos_ols_train_pred)]
```

```python
[metrics.r2_score(bos_y_test, bos_ols_test_pred),
 metrics.mean_squared_error(bos_y_test, bos_ols_test_pred),
 metrics.mean_absolute_error(bos_y_test, bos_ols_test_pred)]
```
